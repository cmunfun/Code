```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 设置matplotlib使用macOS自带的中文字体
plt.rcParams['font.sans-serif'] = ['PingFang SC', 'Arial Unicode MS']
plt.rcParams['axes.unicode_minus'] = False  # 正常显示负号

# 生成随机时间序列数据
np.random.seed(42)
dates = pd.date_range('2024-01-01', periods=30)
values = np.random.randn(30).cumsum()
df = pd.DataFrame({'日期': dates, '数值': values})
display(df.head())

# 折线图（使用彩色线条）
plt.figure(figsize=(10, 5))
plt.plot(df['日期'], df['数值'], marker='o', color='#FF5733', label='数值变化')
plt.title('随机生成的时间序列数据（折线图）')
plt.xlabel('日期')
plt.ylabel('数值')
plt.grid(True)
plt.legend()
plt.show()

# 彩色柱状图
plt.figure(figsize=(10, 5))
colors = plt.cm.rainbow(np.linspace(0, 1, len(df)))
plt.bar(df['日期'], df['数值'], color=colors)
plt.title('随机生成的时间序列数据（彩色柱状图）')
plt.xlabel('日期')
plt.ylabel('数值')
plt.xticks(rotation=45)
plt.show()
```
