```python
import csv
import pandas as pd
```

```python
df = pd.read_csv('/Users/cm/Dataset/appleAppData.csv', nrows=10)
df.head()
```

```python
sql = "CREATE TABLE my_table (\n"
for col in df.columns:
    # 智能推断数据类型
    dtype = 'VARCHAR(255)'
    if pd.api.types.is_numeric_dtype(df[col]):
        dtype = 'DOUBLE'
    elif pd.api.types.is_datetime64_any_dtype(df[col]):
        dtype = 'DATETIME'
    elif len(df[col].dropna()) > 0 and isinstance(df[col].iloc[0], str):
        max_len = df[col].str.len().max()
        dtype = f'VARCHAR({int(max_len * 1.5)})' if not pd.isna(max_len) else 'VARCHAR(255)'
    
    col_clean = col.replace(' ', '_').replace('%', 'pct').lower()  # 清理列名
    sql += f"  `{col_clean}` {dtype},\n"

sql = sql.rstrip(',\n') + "\n);"
```

```python
# 保存SQL文件
with open('create_table.sql', 'w') as f:
    f.write(sql)

print("建表SQL已生成：create_table.sql")
```
